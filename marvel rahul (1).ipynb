```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marvel=pd.read_csv('Marvel.csv')
```

```python
# Assuming 'marvel' is a DataFrame
# Checking for missing values in DataFrame'marvel'
# Counting the number of missing values in each column

marvel.isnull().sum()
```

```python
# Dropping rows with missing values from the DataFrame 'marvel' and creating a new DataFrame 'Marvel' without that rows
Marvel=marvel.dropna()
```

```python
Marvel.isnull().sum()
```

```python
# Checking for duplicate rows in the DataFrame 'marvel'
# Counting the occurance of each unique value (True/False) returned by the duplicated() method
marvel.duplicated().value_counts()

# Dropping duplicate rows from the DataFrame 'marvel' and modifing the DataFrame in place
marvel.drop_duplicates(inplace=True)
```

```python
marvel.shape
```

```python
# Generate decriptive statistics for the numerical columns in the DataFrame
Marvel.describe()
```

```python
# Replace the value of 'Break Even' form 'marvel' DataFrame

Marvel['Break Even'] = Marvel['Break Even'].replace({'Success': 1, 'Flop': 0})
```

```python
Marvel 
```

```python

# Create a new figure with figsize (20,8)
plt.figure(figsize=(20,8))

# Create a scatter plot with 'Film' on the x-axis and 'Rotten Tomatoes Critic Score' on the y-axis
# Each point is colored red and has a size of 100
plt.scatter(Marvel['Film'],Marvel['Rotten Tomatoes Critic Score'], color='red', s=100)

# Set the title of the plot
plt.title('Relation Between Film And Rotten Tomatoes Critic Score', fontsize=20, fontweight='bold')

# Label the x-axis
plt.xlabel('Film', fontsize=18, fontweight='bold')

# Label the y-axis
plt.ylabel('Rotten Tomatoes Critic Score', fontsize=15, fontweight='bold')

# Rotate x-axis labels vertically and align them to the right
plt.xticks(rotation='vertical', ha='right', fontsize=15)

# Set the fontsize of y-axis labels
plt.yticks(fontsize=15)

# Display the plot
plt.show()

```

```python

Film = ['Howard the Duck', 'The Punisher', 'Blade', 'X-Men', 'Blade II', 'Black Panther: wakanda Forever']
box_office_gross_worldwide = [37962774, 61123659, 131211377, 157299718, 139039810, 296339528]

# Set up the figure size
plt.figure(figsize=(12,6))

# Create a bar plot for Film and Box office gross worldwide
plt.bar(Film, box_office_gross_worldwide, color='green')

# Adding labels and title
plt.xlabel('Film', fontsize=12)
plt.ylabel('Box office gross worldwide', fontsize=12)
plt.title('Box Office Gross Worldwide For Marvel Films')


# Show the plot
plt.tight_layout()
plt.show()

```

```python
# Calculate the mean of 'Gross to Budget' column in the Marvel DataFrame
Marvel['Gross to Budget'].mean()
```

```python
# Calculate the median of 'Rotten Tomatoes Critic Score' column in the Marvel DataFrame
Marvel['Rotten Tomatoes Critic Score'].median()
```

```python
# Calculate the standard deviavtion of 'Break Even' column in the Marvel DataFrame
Marvel['Break Even'].std()
```

```python
# Calculate the correlation matrix of numeric columns in the Marvel DataFrame
correlation=Marvel.corr(numeric_only='True')
```

```python
correlation
```

```python
# Set the figuresize
plt.figure(figsize=(25,6))

# Create heatmap using seaborn
sns.heatmap(correlation, annot=True, cmap='RdBu')

# Set the titile for the plot
plt.title('Correlation Matrix', fontsize=20)

# Set the x-axis fontsize
plt.xticks(fontsize=13)

# set the y-axis fontsize
plt.yticks(fontsize=14)

# Display the plot
plt.show()
```
